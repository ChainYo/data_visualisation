## Evaluation sur l'échantillon test

Générons un ensemble de données aléatoirement suivant: `y = 3x + 2`

```python
import matplotlib.pyplot as plt
import numpy as np

X = 4 * np.random.rand(1000,1)
X_b = np.c_[np.ones((1000,1)), X]

Y = 2 + 3 * X + np.random.randn(1000,1)

plt.plot(X,Y,'.')

plt.xlim(0,4)
plt.ylim(0,15)

plt.xlabel(r'x',fontsize=8)
plt.ylabel(r'y',fontsize=8)

plt.title('How to caclulate the mean squared error in  python ?',fontsize=8)

plt.savefig("mean_squared_error_01.png", bbox_inches='tight')

#----- Let's take one random linear model

theta = np.array([[-1.4],[5.0]])

X_new = np.array([[0],[4]])
X_new_b = np.c_[np.ones((2,1)), X_new]

plt.plot(X_new, X_new_b.dot( theta ), '-')

plt.xlim(0,4)
plt.ylim(0,15)

plt.xlabel(r'x',fontsize=8)
plt.ylabel(r'y',fontsize=8)

plt.title('How to caclulate the mean squared error in  python ?',fontsize=8)

plt.savefig("mean_squared_error_02.png", bbox_inches='tight')


Y_predict = X_b.dot( theta )

print(Y_predict.shape, X_b.shape, theta.shape)

mse = np.sum( (Y_predict-Y)**2 ) / 1000.0

print('mse: ', mse)

from sklearn.metrics import mean_squared_error

print('mse (sklearn): ', mean_squared_error(Y,Y_predict))
```

## Validation croisée (k-folds)

<p>Va nous permettre d’utiliser l'intégralité de notre jeu de données pour l’entraînement et pour la validation</p></br>
<p>On découpe le jeu de données en k parties à peu près égales. Tour à tour, chacune des k parties est utilisée comme jeu de test. Le reste (autrement dit, l’union des k-1 autres parties) est utilisé pour l'entraînement.</p>

```python
import numpy as np #importation de numpy
from sklearn.model_selection import KFold #importation de KFold

X = ["a", "b", "c", "d"]
kf = KFold(n_splits=2) # On split les données en 2
for train, test in kf.split(X):
  print("%s %s" % (train, test))
```

## Sélection des variables

Il y a un module **`sklearn.feature_selection`** qui permet de restreindre le choix des features ainsi que la réduction de la dimensionalité sur les ensembles d'échantillons.   
Cela permet l'amélioration de la précision des estimations ainsi que de booster les performances pour des jeux de données très grands.

### Recursive feature elimination

```python
from sklearn.svm import SVC
from sklearn.datasets import load_digits
from sklearn.feature_selection import RFE
import matplotlib.pyplot as plt

# Load the digits dataset
digits = load_digits()
X = digits.images.reshape((len(digits.images), -1))
y = digits.target

# Create the RFE object and rank each pixel
svc = SVC(kernel="linear", C=1)
rfe = RFE(estimator=svc, n_features_to_select=1, step=1)
rfe.fit(X, y)
ranking = rfe.ranking_.reshape(digits.images[0].shape)

# Plot pixel ranking
plt.matshow(ranking, cmap=plt.cm.Blues)
plt.colorbar()
plt.title("Ranking of pixels with RFE")
plt.show()
```

### Recursive feature elimination with cross-validation

```python
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFECV
from sklearn.datasets import make_classification

# Build a classification task using 3 informative features
X, y = make_classification(n_samples=1000, n_features=25, n_informative=3,
                           n_redundant=2, n_repeated=0, n_classes=8,
                           n_clusters_per_class=1, random_state=0)

# Create the RFE object and compute a cross-validated score.
svc = SVC(kernel="linear")
# The "accuracy" scoring is proportional to the number of correct
# classifications
rfecv = RFECV(estimator=svc, step=1, cv=StratifiedKFold(2),
              scoring='accuracy')
rfecv.fit(X, y)

print("Optimal number of features : %d" % rfecv.n_features_)

# Plot number of features VS. cross-validation scores
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score (nb of correct classifications)")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
plt.show()
```

## Définition du principe `SimpleImputer`

Le module `sklearn.impute.SimpleImputer` permet de compléter les valeurs manquantes dans un dataset.  

*Il remplace depuis la version 0.20 `SimpleImputer` du module `sklearn.preprocessing.Imputer` qui a été retiré.*

```python
import numpy as np
from sklearn.impute import SimpleImputer

# On veut remplacer les valeurs absentes par d'autres valeurs
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
# On applique notre module sur le tableau suivant
imp_mean.fit([[7, 2, 3], [4, np.nan, 6], [10, 5, 9]])

X = [[np.nan, 2, 3], [4, np.nan, 6], [10, np.nan, 9]]
print(imp_mean.transform(X))
```

## Définition du principe `LabelEncoder`

Le module `sklearn.preprocessing.LabelEncoder` permet l'encodage des libellés de données cibles.   
Il permet soit de normaliser des libellés numériques, soit des libellés non-numériques.  

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

le.fit([1, 2, 2, 6])

print(le.classes_)
# encodage de [1, 1, 2, 6]
print(le.transform([1, 1, 2, 6]))
# Retour avant l'encodage
print(le.inverse_transform([0, 0, 1, 2]))
```

```python
le = preprocessing.LabelEncoder()
le.fit(["paris", "paris", "tokyo", "amsterdam"])

print(list(le.classes_))

print(le.transform(["tokyo", "tokyo", "paris"]))

print(list(le.inverse_transform([2, 2, 1])))
```

## Définition du principe de Feature Scaling

La standardisation d'un dataset est un pré-requis commun pour plusieurs estimateurs liés au machine learning dans le package Scikit-learn.  


La fonction `scale` permet d'obtenir un aperçu rapide et facile sur un tableau simple :

```python
from sklearn import preprocessing
import numpy as np
X_train = np.array([[ 1., -1.,  2.],
                     [ 2.,  0.,  0.],
                     [ 0.,  1., -1.]])
X_scaled = preprocessing.scale(X_train)

X_scaled
```

Mais ce qui va nous intéresser ici c'est le module `StandardScaler`.  
Il va permettre d'implémenter l'API Transformer pour calculer la moyenne et l'écart-type sur un ensemble d'apprentissage et donc stocker ceci pour pouvoir appliquer la transformation ultérieurement sur l'ensemble des tests.

```python
scaler = preprocessing.StandardScaler().fit(X_train)
print(scaler)

print(scaler.mean_)

print(scaler.scale_)

print(scaler.transform(X_train))
```
